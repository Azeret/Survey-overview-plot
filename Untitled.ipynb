```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad

# Function to convert redshift to approximate age of the universe in billion years
def redshift_to_age(z):
    H0 = 70.0  # Hubble constant in km/s/Mpc
    omega_m = 0.3  # Matter density parameter
    omega_lambda = 0.7  # Dark energy density parameter
    c = 299792.458  # Speed of light in km/s
    # Integrate to find age as a function of redshift
    age = np.zeros_like(z)
    for i, zi in enumerate(z):
        integrand = lambda x: 1.0 / (H0 * np.sqrt(omega_m * (1 + x)**3 + omega_lambda))
        age[i], _ = quad(integrand, zi, np.inf)
    return age / (3600 * 24 * 365.25 * 1e9)  # Convert from seconds to Gyr

# Redshift values for y-axis conversion
z_values = np.linspace(0, 10, 100)
age_values = redshift_to_age(z_values)

# Survey data with colors, duration, and redshift coverage
surveys = {
    "Euclid": {"start_year": 2023, "end_year": 2030, "redshift_min": 0, "redshift_max": 2.0, "color": 'blue'},
    "UVEX": {"start_year": 2030, "end_year": 2040, "redshift_min": 0, "redshift_max": 1.0, "color": 'green'},
    "QUVIK": {"start_year": 2030, "end_year": 2040, "redshift_min": 0, "redshift_max": 1.5, "color": 'red'},
    "LISA": {"start_year": 2034, "end_year": 2040, "redshift_min": 0, "redshift_max": 10, "color": 'cyan'},
    "CE/ET": {"start_year": 2035, "end_year": 2040, "redshift_min": 0, "redshift_max": 10, "color": 'magenta'},
    "Gaia DR3": {"start_year": 2023, "end_year": 2030, "redshift_min": 0, "redshift_max": 0.1, "color": 'orange'},
    "WEAVE": {"start_year": 2023, "end_year": 2030, "redshift_min": 0, "redshift_max": 0.5, "color": 'purple'},
    "4MOST": {"start_year": 2023, "end_year": 2030, "redshift_min": 0, "redshift_max": 0.5, "color": 'brown'},
    "PFS": {"start_year": 2023, "end_year": 2030, "redshift_min": 0, "redshift_max": 0.5, "color": 'pink'},
    "ELT": {"start_year": 2027, "end_year": 2040, "redshift_min": 0, "redshift_max": 3.0, "color": 'grey'}
}

# Create the plot with dual y-axes
fig, ax1 = plt.subplots(figsize=(12, 8))

for survey, info in surveys.items():
    ax1.add_patch(plt.Rectangle((info['start_year'], info['redshift_min']), info['end_year'] - info['start_year'], 
                               info['redshift_max'] - info['redshift_min'], color=info['color'], alpha=0.3, edgecolor='black', linewidth=3))

# Formatting the primary y-axis
ax1.set_xlabel('Year')
ax1.set_ylabel('Redshift Range')
ax1.set_title('Timeline and Redshift Coverage of Astronomical Surveys')
ax1.set_xlim(2020, 2040)
ax1.set_ylim(0, 11)

# Adding secondary y-axis
ax2 = ax1.twinx()
ax2.set_ylabel('Age of Universe (Gyr)')
ax2.set_ylim(redshift_to_age(np.array([0, 11])))
ax2.set_yticks(np.linspace(*ax2.get_ylim(), num=6))  # Set y-ticks to be linear in age

# Setting up the legend
handles = [plt.Rectangle((0, 0), 1, 1, color=info['color'], alpha=0.3, edgecolor='black', linewidth=3) for survey, info in surveys.items()]
ax1.legend(handles, surveys.keys(), loc='upper right')
```
